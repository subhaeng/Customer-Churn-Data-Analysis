```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset (make sure to provide the correct file path)
df = pd.read_csv("Customer Churn.csv")  # Replace with the actual file name

# Group by 'Churn' and count
gb = df.groupby("Churn").agg({'Churn': 'count'})

# Plot pie chart
plt.pie(gb['Churn'], labels=gb.index, autopct='%1.1f%%')
plt.title('Percentage of Churn Distribution')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Get percentage data
grouped = df.groupby(['SeniorCitizen', 'Churn']).size().unstack()
percent = grouped.div(grouped.sum(axis=1), axis=0) * 100  # Convert to %

# Step 2: Plot stacked bar chart
ax = percent.plot(kind='bar', stacked=True, figsize=(4, 4), colormap='Set2')

# Step 3: Add % labels
for i, col in enumerate(percent.columns):
    for j, val in enumerate(percent[col]):
        if val > 0:
            ax.text(j, percent.iloc[:j, i].sum() + val/2, f'{val:.1f}%', 
                    ha='center', va='center', fontsize=10, color='black')

# Final formatting
plt.title("Churn by Senior Citizen Status (Percentage)")
plt.xlabel("Senior Citizen (0 = No, 1 = Yes)")
plt.ylabel("Percentage")
plt.legend(title="Churn", bbox_to_anchor=(0.9,0.9))
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

#compartively a greater percentage of people in senior category have churned

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example: loading the dataset
df = pd.read_csv("Customer Churn.csv")  # Replace with your actual file path

# List of categorical columns you want to plot
columns_to_plot = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Set up the subplot grid (3 rows, 3 columns for 9 plots)
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
fig.suptitle("Count Plots for Service Columns", fontsize=16)

# Flatten axes for easy iteration
axes = axes.flatten()

# Loop through columns and create plots
for i, col in enumerate(columns_to_plot):
    sns.countplot(data=df, x=col, ax=axes[i],hue=df["Churn"])
    axes[i].set_title(col)
    axes[i].tick_params(axis='x', rotation=45)
    axes[i].bar_label(axes[i].containers[0])  # Add labels on bars

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```

#Customers without services like OnlineSecurity, TechSupport, and DeviceProtection show noticeably higher churn rates, suggesting these features might influence retention.
